# SAVE PRESSURE NORMALIZED DATA FUNCTION


```python
# SAVE PRESSURE NORMALIZED DATA 
def save_pressure_normalized_data(folder_file, range1, range2, added_limit, norm_type, window_size, directory):
    all_subject_pressure_data = pd.DataFrame()
    all_subject_accel_data = pd.DataFrame()
    # ITERATE EACH SUBJECT
    # file = 0
    # while (file != 2):
#     directory = '/media/togzhan/windows 2/documents/lump_files/data/trial_records/data_collection/'
    for file in folder_file:#folder_file:
#         file = 'Asset'
        single_subject_pressure_data = pd.DataFrame()
        single_subject_accel_data = pd.DataFrame()
        csv_files = []
        directory_folder = directory + file #folder_file[file] #
        os.chdir(directory_folder)
    #     print(directory_folder)
        object_folders = glob.glob('*')
        print(object_folders)
        # ITERATE EACH OBJECT
        for i in range(range1, range2): #13
            single_folder_pressure_data = pd.DataFrame()
            single_folder_accel_data = pd.DataFrame()
            os.chdir(directory_folder + '/' + str(i + 1))
            csv_files = glob.glob('*.csv')
            print('Folder: ', directory_folder + '/' + str(i + 1), '\n')   
    #         print(csv_files)
            # ITERATE EACH FILE
            skipped = 0
            added = 0;
            for j in range(len(csv_files)): # len(csv_files)
                single_ed_pressure_data = pd.DataFrame()
                single_pressure_data = pd.DataFrame()
                single_accel_data = pd.DataFrame()
                single_csv_data = pd.read_csv(csv_files[j], header = None)
                single_csv_data.columns = [''] * len(single_csv_data.columns)
                single_pressure_data = pd.DataFrame(single_csv_data.values[1:1121, 1:31]) 
    #             single_accel_data = pd.DataFrame(single_csv_data.values[1:, 31:])
                # ITERATE EACH SENSOR
    #             print('CONDITION: ', added == added_limit, '\n')
                if((len(single_pressure_data) < 1120) or (added == added_limit)):
                    skipped = skipped + 1
                    print('File skipped: ', csv_files[j], '\n')
#                     break
                    continue
    
                for k in range(15):
                    single_sensor_data = pd.DataFrame()
                    single_sensor_data = pd.DataFrame(single_pressure_data.values[:, finger_sensors[k]])
                    
                    if (norm_type == 0):
                        single_sensor_data_norm = normalize_mode1(single_sensor_data)
                    elif(norm_type == 1):
                        single_sensor_data_norm = normalize_mode2(single_sensor_data, window_size)
                    elif(norm_type == 2):
                        single_sensor_data_norm = normalize_mode3(single_sensor_data, window_size)
#                     plt.plot(single_sensor_data_norm.values[:])
#                     plt.show()
                    else:
                        single_sensor_data_norm = single_sensor_data
                
                    single_ed_pressure_data = pd.concat([single_ed_pressure_data, single_sensor_data_norm], axis = 0)
                
                single_folder_pressure_data = pd.concat([single_folder_pressure_data, single_ed_pressure_data], axis = 1)
            
                added = added + 1;

            if(np.shape(single_folder_pressure_data)[1] != added_limit):
                print('FOLDER SKIPPED: ', np.shape(single_folder_pressure_data), '\n')
#                 break
                continue
            single_subject_pressure_data = pd.concat([single_subject_pressure_data, single_folder_pressure_data], axis = 1) 
        all_subject_pressure_data = pd.concat([all_subject_pressure_data, single_subject_pressure_data], axis = 1)            
    #     file = file + 1        
    return all_subject_pressure_data
```

# NORMALIZATION FUNCTIONS

```python
def normalize_mode1(single_sensor_data):
    #1120 by 15
    single_sensor_data_norm = []
    max_abs_scaler = preprocessing.MinMaxScaler()
    single_sensor_data_norm = max_abs_scaler.fit_transform(single_sensor_data)
    return pd.DataFrame(single_sensor_data_norm)

def normalize_mode2(single_sensor_data, window_size):
    i = 0
    single_sensor_data_norm = []
    size = len(single_sensor_data)/window_size 
    if((len(single_sensor_data) % window_size) != 0):
        print('Invalid size\n')
    while(i != size):
        single_sensor_data_norm.append(np.std((single_sensor_data.values[i*window_size:(i+1)*window_size])))
        i = i + 1
    return pd.DataFrame(single_sensor_data_norm)

def normalize_mode3(single_sensor_data, window_size):
    i = 0
    single_sensor_data_norm = []
    size = len(single_sensor_data)/window_size
    if((len(single_sensor_data) % window_size) != 0):
        print('Invalid size\n')
    while(i != size):
        single_sensor_data_norm.append(np.mean((single_sensor_data.values[i*window_size:(i+1)*window_size])))
        i = i + 1
    return pd.DataFrame(single_sensor_data_norm)
    
```

```python
directory = '/Volumes/Samsung_T5/project files/lump_project/data/data_collection/W_all_data/normalized'
os.chdir(directory)
csv_files = glob.glob('all_*0_1.csv')

all_subject_pressure_data_with_0_1 = pd.read_csv(csv_files[1], header = None)
all_subject_pressure_data_with_0_1 = all_subject_pressure_data_with_0_1.transpose()

all_subject_pressure_data_without_0_1 = pd.read_csv(csv_files[0], header = None)
all_subject_pressure_data_without_0_1 = all_subject_pressure_data_without_0_1.transpose()
```

```python
all_with_std = pd.DataFrame()
all_without_std = pd.DataFrame()
all_with_mean = pd.DataFrame()
all_without_mean = pd.DataFrame()

all_with_std_ = pd.DataFrame()
all_without_std_ = pd.DataFrame()
all_with_mean_ = pd.DataFrame()
all_without_mean_ = pd.DataFrame()

for i in range(np.shape(all_subject_pressure_data_with_0_1)[0]):
    values = all_subject_pressure_data_with_0_1.values[:]
    values_without = all_subject_pressure_data_without_0_1.values[:]
    all_with_std = pd.DataFrame()
    all_without_std = pd.DataFrame()
    all_with_mean = pd.DataFrame()
    all_without_mean = pd.DataFrame()
    for j in range(15):        
        all_with_std = np.append(all_with_std, np.std(values[i, j*1120:(j+1)*1120]))
        all_without_std = np.append(all_without_std, np.std(values_without[i, j*1120:(j+1)*1120]))
        
        all_with_mean = np.append(all_with_mean, np.mean(values[i, j*1120:(j+1)*1120]))
        all_without_mean= np.append(all_without_mean, np.mean(values_without[i, j*1120:(j+1)*1120]))
        
    all_with_std = pd.DataFrame(all_with_std).transpose()
    all_with_std = all_with_std.reset_index(drop = True)
    all_without_std = pd.DataFrame(all_without_std).transpose()
    all_without_std = all_without_std.reset_index(drop = True)
    
    all_with_mean = pd.DataFrame(all_with_mean).transpose()
    all_with_mean = all_with_mean.reset_index(drop = True)
    all_without_mean = pd.DataFrame(all_without_mean).transpose()
    all_without_mean = all_without_mean.reset_index(drop = True)
    
        
    all_with_std_ = pd.concat([all_with_std_, pd.DataFrame(all_with_std)], axis = 0)
    all_without_std_ = pd.concat([all_without_std_, pd.DataFrame(all_with_std)], axis = 0)
    all_with_mean_ = pd.concat([all_with_mean_, pd.DataFrame(all_with_std)], axis = 0)
    all_without_mean_ = pd.concat([all_without_mean_, pd.DataFrame(all_with_std)], axis = 0)
    
directory = '/Volumes/Samsung_T5/project files/lump_project/data/data_collection/W_all_data/'
os.chdir(directory)
all_with_std_.to_csv('all_subject_pressure_data_with_std_norm.csv', index = False, header = False)
all_without_std_.to_csv('all_subject_pressure_data_without_std_norm.csv', index = False, header = False)
all_with_mean_.to_csv('all_subject_pressure_data_with_mean_norm.csv', index = False, header = False)
all_without_mean_.to_csv('all_subject_pressure_data_without_mean_norm.csv', index = False, header = False)
```
