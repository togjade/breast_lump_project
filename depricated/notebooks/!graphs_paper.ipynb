```python
def plot_accuracies(df, name):
    fig = go.Figure()
    durations = df['Duration'].unique()
    bar_width = 0.4  # Adjust this value as needed for bar width
    fig.add_trace(go.Bar(
        x=durations,
        y=df[df['Configuration'] == 'Initial']['Lump Accuracy'],
        name='Initial Lump Accuracy',
        marker_color='green',
        marker_pattern_shape=""
#         width=[bar_width] * len(durations)
    ))

    fig.add_trace(go.Bar(
        x=durations,
        y=df[df['Configuration'] == 'Fine-tuned']['Lump Accuracy'],
        name='Fine-tuned Lump Accuracy',
        marker_color='yellow',
        marker_pattern_shape=""
    ))

    fig.add_trace(go.Bar(
        x=durations,
        y=df[df['Configuration'] == 'Initial']['Size Accuracy'],
        name='Initial Size Accuracy',
        marker_color='green', 
        marker_pattern_shape="/"
    ))

    fig.add_trace(go.Bar(
        x=durations,
        y=df[df['Configuration'] == 'Fine-tuned']['Size Accuracy'],
        name='Fine-tuned Size Accuracy',
        marker_color='yellow',
        marker_pattern_shape="/"
    ))

    fig.add_trace(go.Bar(
        x=durations,
        y=df[df['Configuration'] == 'Initial']['Position Accuracy'],
        name='Initial Position Accuracy',
        marker_color='green',
        marker_pattern_shape="\\"
    ))

    fig.add_trace(go.Bar(
        x=durations,
        y=df[df['Configuration'] == 'Fine-tuned']['Position Accuracy'],
        name='Fine-tuned Position Accuracy',
        marker_color='yellow',
        marker_pattern_shape="\\"
    ))

    fig.update_layout(
        barmode='group',
        xaxis_tickangle=-0,
        xaxis=dict(title='Duration'),
        yaxis=dict(title='Accuracy', range=[0.5, 1.0]),
        legend=dict(x=1.05, y=1.0),
        font=dict(
            size=24,
            color='Black',
        ),
        )

    fig.show()
    fig.write_image(name)
```

```python
# Separate the configurations for plotting
df_distal = df[:14]
df_middle = df[14:28]
df_all = df[28:]
```

```python
def plot_accuracies(df, row, fig, show_legend=False):
    durations = df['Duration'].unique()
    yaxis_name = f'y{row+1}'
    fig.add_trace(go.Bar(
        x=durations,
        y=df[df['Configuration'] == 'Initial']['Lump Accuracy'],
        name='Baseline: Lump Presence' if show_legend else None,
        marker_color='green',
        marker_pattern_shape="",
        yaxis=yaxis_name,
        showlegend=show_legend
    ))

    fig.add_trace(go.Bar(
        x=durations,
        y=df[df['Configuration'] == 'Fine-tuned']['Lump Accuracy'],
        name='New Data: Lump Presence' if show_legend else None,
        marker_color='yellow',
        marker_pattern_shape="",
        yaxis=yaxis_name,
        showlegend=show_legend
    ))

    fig.add_trace(go.Bar(
        x=durations,
        y=df[df['Configuration'] == 'Initial']['Size Accuracy'],
        name='Baseline: Size' if show_legend else None,
        marker_color='green',
        marker_pattern_shape="/",
        yaxis=yaxis_name,
        showlegend=show_legend
    ))

    fig.add_trace(go.Bar(
        x=durations,
        y=df[df['Configuration'] == 'Fine-tuned']['Size Accuracy'],
        name='New Data: Size' if show_legend else None,
        marker_color='yellow',
        marker_pattern_shape="/",
        yaxis=yaxis_name,
        showlegend=show_legend
    ))

    fig.add_trace(go.Bar(
        x=durations,
        y=df[df['Configuration'] == 'Initial']['Position Accuracy'],
        name='Baseline: Position' if show_legend else None,
        marker_color='green',
        marker_pattern_shape="\\",
        yaxis=yaxis_name,
        showlegend=show_legend
    ))

    fig.add_trace(go.Bar(
        x=durations,
        y=df[df['Configuration'] == 'Fine-tuned']['Position Accuracy'],
        name='New Data: Position' if show_legend else None,
        marker_color='yellow',
        marker_pattern_shape="\\",
        yaxis=yaxis_name,
        showlegend=show_legend
    ))
```

```python
fig = go.Figure()

# Separate the configurations for plotting
df_distal = df[:14]
df_middle = df[14:28]
df_all = df[28:]

# Plot each dataframe
plot_accuracies(df_distal, 0, fig, show_legend=True)
plot_accuracies(df_middle, 1, fig, show_legend=False)
plot_accuracies(df_all, 2, fig, show_legend=False)

# Define y-axes explicitly
fig.update_layout(
    yaxis=dict(
        title='',
        domain=[0, 0.32],
        range=[0.5, 1.0]
    ),
    yaxis2=dict(
        title='Accuracy',
        domain=[0.34, 0.66],
        range=[0.5, 1.0]
    ),
    yaxis3=dict(
        title='',
        domain=[0.68, 1.0],
        range=[0.5, 1.0]
    )
)
# Update the layout for the combined plot
fig.update_layout(
    height=1000,  # Adjust the height as needed
    xaxis_title='Duration',
    legend=dict(
        x=0.5,
        y=-0.1,
        xanchor='center',
        font=dict(size=28),
        orientation='h'
    ),
    font=dict(size=28, color='Black'),
    showlegend=True,
    barmode='group',
    margin=dict(t=50, b=100),
    # Adjust top and bottom margins as needed
)

fig.add_annotation(
    text="First Row",
    xref="paper", yref="paper",
    x=0.5, y=0.279,
    showarrow=False,
    font=dict(size=28)
)

fig.add_annotation(
    text="First and Second Row",
    xref="paper", yref="paper",
    x=0.5, y=0.640,
    showarrow=False,
    font=dict(size=28)
)

fig.add_annotation(
    text="All Rows",
    xref="paper", yref="paper",
    x=0.5, y=1,
    showarrow=False,
    font=dict(size=28)
)

fig.show()
fig.write_image("combined_accuracies.pdf")
```
