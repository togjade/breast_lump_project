```python
# choose the directory and necessary files



def plot_pressure_once(df):

    col_type = df[df[df.columns[1]] == 0]
    print(df.columns[1])

    
    print(col_type.columns[3])
    col_data = col_type['Data']
    c = col_data
    # c = c.values[:]

    fig = make_subplots()
    i = 0
    fig.update_layout(
        {
        "plot_bgcolor": "rgba(0, 0, 0, 0)",
        "paper_bgcolor": "rgba(0, 0, 0, 0)",
        },
        font=dict(
            size=32,
            ),
        
        )
    # fig.update_xaxes(showgrid=False, zeroline=False)
    # fig.update_yaxes(showgrid=False, zeroline=False)
    for row in range(5):
        for col in range(3):
            d = c[15][i*1120:(i+1)*1120]
            # d = d - np.mean(d)
            x = pd.DataFrame(d).index.values.tolist()
            fig.add_trace(
                go.Scatter(x=x, y=d, name=f"S {i+1}"),
                row=1, col=1
            )
            i = i+1
    fig.update_layout(height=800, width=1300)
    fig.update_traces(line={'width': 8})

    return fig
```

```python
data = pd.DataFrame(data)
fig = plot_pressure_once(data)
fig.show()    
fig.write_image("pressure_units.pdf")
```

```python
def plot_pressure_data2(df, row_idx=0, num_segments=15):
    """
    Plot pressure data from CSV format (data2)
    df: DataFrame with numeric columns (wide format)
    row_idx: which row to plot
    num_segments: how many segments to split the data into (default 15 for 5x3 grid)
    """
    # Get the row data (all columns)
    row_data = df.iloc[row_idx].values
    
    # Calculate segment length
    total_length = len(row_data)
    segment_length = total_length // num_segments
    
    fig = make_subplots(rows=5, cols=3)
    i = 0
    for row in range(5):
        for col in range(3):
            if i < num_segments:
                d = row_data[i*segment_length:(i+1)*segment_length]
                x = list(range(len(d)))
                fig.add_trace(
                    go.Scatter(x=x, y=d),
                    row=row+1, col=col+1
                )
            i = i+1
    
    fig.update_layout(height=600, width=1300,
                      title_text=f"Pressure Data - Row {row_idx} (15 segments)")
    
    return fig
```

```python
# Plot data2 (you can change row_idx to see different rows)
fig2 = plot_pressure_data2(data2, row_idx=15)
fig2.show()
```
